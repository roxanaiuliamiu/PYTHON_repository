```python
#Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day”
#to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single
#busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.
```

```python
busiest_days= []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value ==1:
    busiest_days.append("Busiest days")
  elif value == 4 or value==3:
    busiest_days.append("Slow days")
  else:
    busiest_days.append("Regularly busy")
```

```python
df_ords_prods_merge['busiest_days']=busiest_days
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
#When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you
#to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” 
#and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”
```

```python
#finding the frequency of hour_of_order
```

```python
df_ords_prods_merge['hour_of_order'].value_counts(dropna=False)
```

```python
result_hours = []
for value in df_ords_prods_merge["hour_of_order"]:
  if value >= 10 and value <=16:
    result_hours.append("Most orders")
  elif value >=0 and value <=5:
    result_hours.append("Fewest orders")
  else:
    result_hours.append("Average orders")
```

```python
df_ords_prods_merge['busiest_period_of_day']=result_hours
```
